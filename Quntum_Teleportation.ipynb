```python
import pennylane as qml
import matplotlib.pyplot as plt
import numpy as np
```

###4 Steps are needed:

1.   Prepare the state to teleport
2.   Using Entanglement as a shared resource :Producing Bell's state
3.   Change the basis
4.   Measurement





image.png

###STATE PREPARATION :

```python
def state_preparation():
    qml.Hadamard(wires=0)
    qml.Rot(0.1, 0.2, 0.3, wires=0)
    return qml.state()
```

###ENTAGLEMENT:

```python
def entangle_qubits():
    # ENTANGLE THE SECOND QUBIT (WIRES=1) AND THE THIRD QUBIT
    qml.Hadamard(wires=1)
    qml.CNOT(wires=[1,2])
```

###CHANGE OF BASIS:

```python
def change_basis():
    qml.CNOT(wires=[0,1])
    qml.Hadamard(wires=0)
```

###MEASUREMENT :

```python
def measure():
  qml.CNOT(wires=[1,2])
  qml.CZ(wires=[0,2])
```

###FINAL ALGORITHM:

```python
dev = qml.device("default.qubit", wires=3)
@qml.qnode(dev)
def teleportation():
    state_preparation()
    entangle_qubits()
    change_basis()
    measure()
    state=qml.state()
    return np.array([2*state[0],2*state[1]])
```
